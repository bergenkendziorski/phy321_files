So far we've covered both numerical and analytical solutions for solving physics problems (particularly differential equations). One of the main takeaways from what we've covered so far is that from a force you can find a general solution, but you need initial conditions (typically initial position and velocity) to be able to find a particular solution. These particular solutions can vary quite dramatically, and for the same force can range anywhere from interesting chaotic motion to just blowing up to infinity.

One of the first things we did after some mathematical background was modeling situations by drawing force diagrams. We also distinguished between fundamental forces (eg gravity) and modeled forces with inherent inaccuracy/uncertainty (eg air resistance), as well as internal vs external forces.

On the numerical side of things, we've covered 4 main methods: euler's method, euler-cromer, velocity-verlet, and more recently we're started covering Runge-Kutte (both RK2 and RK4). The euler-cromer and velocity-verlet methods are noteworthy in that they conserve energy, so are very useful in modeling situations where you have a force that conserves energy. To determine if a force conserves energy, you can take the curl of the force (cross product between the del operator and the force) and if that equals 0 then the force is conservative. Conservative forces essentially means that the work done in moving a particle from one position to another is independent of the path, and depends only on the initial and final locations. A sign that a force might be conservative is if it depends only on the position (has no velocity dependence).

Along with conservation of energy, we've also looked at conservation of linear and angular momentum. Linear momentum is defined as simply p=m*v, and is conserved if the derivative equals 0. Often times we assume the mass to be constant with time. We've found that due to Newton's third law in many of our cases with two-body interacting forces, the internal forces cancel and you're left with conservation of linear momentum if there's no external force. Angular momentum is defined as the cross product between position and momentum (r x p) and is often times rewritten as mr x v. Again using invariance of mass with time, the derivative of angular momentum in time can be written as the cross product between the position and the force (r x F). This change in angular momentum is called torque, and if the torque is zero then your system conserves angular momentum.

On the analytical side, the primary approach we've been taking is using the basic F=ma to find acceleration, and then integrating up using initial conditions. We've learned some basic things to look for that might signal a problem can't be solved analytically, typically that you have multiple dimensions that have differential equations depending on each other. In these situations we'd use an appropriate numerical method described above, although sometimes the coordinate system can be changed (often times to spherical coordinates) to create equations that can be solved analytically.

A major topic is connecting all of these different parts. Often we look at the potential V, which is related to the force since the force equals the negative gradient of the potential. A graph of the potential can often times be helpful in analyzing the motion of a system, as you can see stable and unstable equilibrium points, as well as where potential is gained or lost. In many of these cases, this increase or decrease in potential can be directly converted to a corresponding change in kinetic energy (1/2*mv^2), so it's quite easy to visualize how a particle with a given total energy would behave given a specific initial position.

Recently we've been covering harmonic oscillations, which are important because it's a model than can be used as an approximation for many physical scenarios. Besides the obvious mass on a spring or the pendulum, harmonic oscillations can approximate a mass stuck in a potential energy well, and comes up routinely in other physics situations (eg quantum mechanics). Our general equation for harmonic oscillation is m(d^2x/dt^2) + b(dx/dt) + kx(t) = F_0*cos(wt). In this situation we have a driving force of amplitude F_0, which we've seen to often dominate after a sufficient time has passed. In many cases we set F_0 to 0, and can end with 3 scenarios (underdamping, critical damping, and overdamping, as defined by parameter gamm = b/(2*m*w_0) where w_0=sqrt(k/m). In all of these cases we have the damping term of b(dx/dt), and while we've studied the harmonic oscillator without this term, it's not nearly as interesting.

4c) I'll use the Runge-Kutte 4 method (RK4) since it's the most numerically precise of the method's that we've studied. It doesn't conserve energy, which is fine since we have a damping term anyway (Runge-Kutte is good when we have a force that causes our system to lose energy).

```python
## 4c) #The first 2 functions are taken from class notes.
%matplotlib inline
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt

#these I'm defining as global variables so that I can easily put everything into functions (makes later parts easier)
DeltaT = 0.001
tfinal = 20 # in dimensionless time
n = ceil(tfinal/DeltaT)
gamma = 0.2
Omegatilde = 0.5
Ftilde = 1.0

def RK4(v,x,t,n,Force):
    for i in range(n-1):
# Setting up k1
        k1x = DeltaT*v[i]
        k1v = DeltaT*Force(v[i],x[i],t[i])
# Setting up k2
        vv = v[i]+k1v*0.5
        xx = x[i]+k1x*0.5
        k2x = DeltaT*vv
        k2v = DeltaT*Force(vv,xx,t[i]+DeltaT*0.5)
# Setting up k3
        vv = v[i]+k2v*0.5
        xx = x[i]+k2x*0.5
        k3x = DeltaT*vv
        k3v = DeltaT*Force(vv,xx,t[i]+DeltaT*0.5)
# Setting up k4
        vv = v[i]+k3v
        xx = x[i]+k3x
        k4x = DeltaT*vv
        k4v = DeltaT*Force(vv,xx,t[i]+DeltaT)
# Final result
        x[i+1] = x[i]+(k1x+2*k2x+2*k3x+k4x)/6.
        v[i+1] = v[i]+(k1v+2*k2v+2*k3v+k4v)/6.
        t[i+1] = t[i] + DeltaT

def SpringForce(v,x,t): #returns acceleration
    return  -2*gamma*v-x+Ftilde*cos(t*Omegatilde)

def run_rk4():
    # set up arrays for t, v, and x
    t = np.zeros(n)
    v = np.zeros(n)
    x = np.zeros(n)

    x0 =  1.0 
    v0 = 0.0
    x[0] = x0
    v[0] = v0
    
    # Start integrating using the RK4 method
    # Note that we define the force function as a SpringForce
    RK4(v,x,t,n,SpringForce)
    KE = .5*v*v
    PE = .5*x*x
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.plot(t, x)
    ax2.plot(t, KE, color = 'r', label = 'kinetic')
    ax2.plot(t, PE, color = 'b', label = 'potential')
    ax2.plot(t, KE+PE, color = 'g', label = 'total')
    #plt.ylabel('x[m]')
    #plt.xlabel('t[s]')
    if Ftilde != 0:
        fig.suptitle("Driven")
    elif gamma < 1:
        fig.suptitle("Under damping")
    elif gamma == 1:
        fig.suptitle("Critical damping")
    else:
        fig.suptitle("Over damping")
        
    ax1.set_ylabel('x[m]')
    ax1.set_xlabel('t[s]')
    ax2.set_ylabel('energy')
    ax2.set_xlabel('t[s]')
    fig.tight_layout()
    fig.legend()
    plt.show()
```

```python
#for this part, we have no driving force
Ftilde = 0.0

#now we consider the 3 scenarios (under, over, critical damping)
gamma = 0.1
run_rk4()
gamma = 1.0
run_rk4()
gamma = 3.0
run_rk4()
```

All of the numerical results are exactly what we would expect. In the case of under damping (gamma < 1) we get oscillatory motion that follows the general trend of exponential decay in the magnitude of the extremes over time (We can also see this general decay through the total energy over time). The more we decrease gamma, the more oscillatory motion we see. With critical damping (gamma = 1) we see a very rapid drop down towards x = 0. In the case of over damping (gamma > 1) we see a much slower exponential decay down towards x = 0. The farther above 1 gamma is, the slower the decay is, which is what we would have expected from our analytical results. With critical damping, you can see that the kinetic energy has a larger spike thant the overdamping case, which makes sense because it accelerates more quickly towards 0. Overall, the numerical results and the analytic results align perfectly.
