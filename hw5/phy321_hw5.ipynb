### Exercise 5 (10pt), Warming up for the Earth-Sun system, Scaling the Equations

The aim of this exercise (as well as the next) is to study the motion
of objects under the influence of the gravitational force.  We will
limit ourselves to the Earth-Sun system. Here we will scale the
equations and sketch our first algorithm for solving the equations,
namely using Euler's method again, as we did in homework sets 3 and 4.  This part
together with the numerical part forms also the entry point for the
solar system honors project. Furthermore, we will reuse parts of these
results when analyzing energy conservation in homework 6.

We will limit ourselves (in order to test the algorithm) to a
hypothetical solar system with the Earth only orbiting around the sun.
The only force in the problem is gravity. Newton's law of gravitation
is given by a force $F_G$

$$
F_G=\frac{GM_{\odot}M_{\mathrm{Earth}}}{r^2},
$$

where $M_{\odot}$ is the mass of the Sun and $M_{\mathrm{Earth}}$ is
the mass of the Earth. The gravitational constant is $G$ and $r$ is
the distance between the Earth and the Sun.  The Sun
has a mass which is much larger than that of the Earth. We can
therefore safely neglect the motion of the Sun in this problem.

We assume that the orbit of the Earth around the Sun 
is co-planar, and we take this to be the $xy$-plane.
Using Newton's second law of motion we get the following equations

$$
\frac{d^2x}{dt^2}=\frac{F_{G,x}}{M_{\mathrm{Earth}}},
$$

and

$$
\frac{d^2y}{dt^2}=\frac{F_{G,y}}{M_{\mathrm{Earth}}},
$$

where $F_{G,x}$ and $F_{G,y}$ are the $x$ and $y$ components of the
gravitational force.

We will use so-called astronomical units when rewriting our equations.
Using astronomical units (AU as abbreviation)it means that one
astronomical unit of length, known as 1 AU, is the average distance
between the Sun and Earth, that is $1$ AU = $1.5\times 10^{11}$ m.  It
can also be convenient to use years instead of seconds since years
match better the time evolution of the solar system. The mass of the
Sun is $M_{\mathrm{sun}}=M_{\odot}=2\times 10^{30}$ kg. The mass of
Earth is $M_{\mathrm{Earth}}=6\times 10^{24}$ kg with a distance 1AU.

In setting up the equations we limit ourselves to a co-planar
motion and use only the $x$ and $y$ coordinates. But you should feel
free to extend your equations to three dimensions, it is not very
difficult and the data from NASA are all in three dimensions.

[NASA](http://www.nasa.gov/index.html) has an excellent site at <http://ssd.jpl.nasa.gov/horizons.cgi#top>.
From there you can extract initial conditions in order to start your differential equation solver.
At the above website you need to change from **OBSERVER** to **VECTOR** and then write in the planet you are interested in.
The generated data contain the $x$, $y$ and $z$ values as well as their corresponding velocities. The velocities are in units of AU per day.
Alternatively they can be obtained in terms of km and km/s. 

For the system below involving only the Earth and the Sun, you
could just initialize the position with say $x=1$ AU and $y=0$ AU.

We assume that mass units can be obtained by using the fact that
Earth's orbit is almost circular around the Sun.

For circular motion we know that the force must obey the following relation

$$
F_G= \frac{M_{\mathrm{Earth}}v^2}{r}=\frac{GM_{\odot}M_{\mathrm{Earth}}}{r^2},
$$

where $v$ is the velocity of Earth. 
The latter equation can be used to show that

$$
v^2r=GM_{\odot}=4\pi^2\mathrm{AU}^3/\mathrm{yr}^2.
$$

Here we add a code-example which may aid in the above studies using Euler's forward method.

I'll start with the standard Euler method, since the general code for that was already provided.

```python
DeltaT = 0.001
#set up arrays 
tfinal = 10 # in years
n = ceil(tfinal/DeltaT)
# set up arrays for time t, velocity v, and position r
t = np.zeros(n)
v = np.zeros((n,2))
r = np.zeros((n,2))
# Initial conditions as compact 2-dimensional arrays
r0 = np.array([1.0,0.0])
v0 = np.array([0,2*pi]) #We can alter this velocity a bit, but since we're starting at [1,0], we need it close to this to
                           #maintain a circular orbit
r[0] = r0
v[0] = v0
Fourpi2 = 4*pi*pi
# Start integrating using Euler's method
for i in range(n-1):
    # Set up the acceleration
    # Here you could have defined your own function for this
    rabs = sqrt(sum(r[i]*r[i]))
    a =  -Fourpi2*r[i]/(rabs**3)
    # update velocity, time and position using Euler's forward method
    v[i+1] = v[i] + DeltaT*a
    r[i+1] = r[i] + DeltaT*v[i]
    t[i+1] = t[i] + DeltaT

plt.title("standard euler method")
plt.plot(r[:,0],r[:,1])
plt.xlabel("x position")
plt.ylabel("y position")
plt.show()

plt.title("showing how the orbit changes")
plt.plot(t,r[:,0])
plt.xlabel("time")
plt.ylabel("x position")
plt.show()
```

We can see that this doesn't work particularly well. It maintains an orbit, but the orbit is not stable. From the second image we can see that the earth is continually getting farther and farther from the sun (not good!). We can somewhat fix this by switching to the euler-cromer method, so I'll do that below.

```python
#Euler-cromer method

def eulercromer(DeltaT = .001):
    #set up arrays 
    tfinal = 10 # in years
    n = ceil(tfinal/DeltaT)
    # set up arrays for time t, velocity v, and position r
    t = np.zeros(n)
    v = np.zeros((n,2))
    r = np.zeros((n,2))
    # Initial conditions as compact 2-dimensional arrays
    r0 = np.array([1.0,0.0])
    v0 = np.array([0,2*pi]) #We can alter this velocity a bit, but since we're starting at [1,0], we need it close to this to
                               #maintain a circular orbit
    r[0] = r0
    v[0] = v0
    Fourpi2 = 4*pi*pi
    # Start integrating using Euler-Cromer's method
    for i in range(n-1):
        # Set up the acceleration
        # Here you could have defined your own function for this
        rabs = sqrt(sum(r[i]*r[i]))
        a =  -Fourpi2*r[i]/(rabs**3)
        # update velocity, time and position using Euler's forward method
        v[i+1] = v[i] + DeltaT*a
        r[i+1] = r[i] + DeltaT*v[i+1]
        t[i+1] = t[i] + DeltaT
    return t,v,r

t, v, r = eulercromer()
plt.title("euler-cromer method")
plt.plot(r[:,0],r[:,1])
plt.xlabel("x position")
plt.ylabel("y position")
plt.show()
```

Already this visually looks much better, but now let's see how much we can improve the euler-cromer method by varying the timestep.

```python
timestep = np.arange(.1, .001, -.001)
maxes = np.zeros(len(timestep))
ctr = 0
for ts in timestep:
    t, v, r = eulercromer(ts)
    maxes[ctr] = max(r[:,0] - 1)
    ctr+=1
plt.plot(timestep,maxes)
plt.title("euler-cromer varying timesteps")
plt.ylabel("max x-distance over starting 1")
plt.xlabel("time step")
```

The above code varies the timestep and is a relatively crude way of showing that the euler-cromer method no longer functions properly above around a time step of 0.08 years. In these larger timesteps, we can see the circular orbit gets off track over time, since the x-position shouldn't be getting above 1.0 AU


Now, we can move on to the velocity-verlet method

```python
#velocity-verlet method

def velocityverlet(DeltaT = .001):
    #set up arrays 
    tfinal = 10 # in years
    n = ceil(tfinal/DeltaT)
    # set up arrays for time t, velocity v, and position r
    t = np.zeros(n)
    v = np.zeros((n,2))
    r = np.zeros((n,2))
    # Initial conditions as compact 2-dimensional arrays
    r0 = np.array([1.0,0.0])
    v0 = np.array([0,2*pi]) #We can alter this velocity a bit, but since we're starting at [1,0], we need it close to this to
                               #maintain a circular orbit
    r[0] = r0
    v[0] = v0
    Fourpi2 = 4*pi*pi
    # Start integrating using velocity verlet method
    for i in range(n-1):
        # Set up the acceleration
        # Here you could have defined your own function for this
        rabs = sqrt(sum(r[i]*r[i]))
        a =  -Fourpi2*r[i]/(rabs**3)
        # update velocity, time and position using velocity-verlet method
        r[i+1] = r[i] + DeltaT*v[i] + a*DeltaT*DeltaT/2
        rabs = sqrt(sum(r[i+1]*r[i+1]))
        a_next = -Fourpi2*r[i+1]/(rabs**3)
        v[i+1] = v[i] + (DeltaT/2)*(a_next+a)
    return t,v,r
```

```python
t, v, r = velocityverlet()
plt.title("velocity-verlet method")
plt.plot(r[:,0],r[:,1])
plt.xlabel("x position")
plt.ylabel("y position")
plt.show()
```

Right away we can see that the velocity-verlet method works very well, but let's test it with varying timesteps so that we can truly compare it to the euler-cromer method.

```python
timestep = np.arange(.1, .001, -.001)
maxes = np.zeros(len(timestep))
ctr = 0
for ts in timestep:
    t, v, r = velocityverlet(ts)
    maxes[ctr] = max(r[:,0] - 1)
    ctr+=1
plt.plot(timestep,maxes)
plt.title("velocity verlet varying timesteps")
plt.ylabel("max x-distance over starting 1")
plt.xlabel("time step")
```

We can see that the velocity-verlet works better than the euler-cromer method with larger time steps, since we're still only off by a max of 0.175 with a time step of 0.10 years, as opposed to around 0.35 years.

### MOVING ON TO 6B) ###

Now let's look for a velocity to escape the sun. I'll use the velocity-verlet method since that's a standard.

```python
#set up arrays 
tfinal = 50 # in years
n = ceil(tfinal/DeltaT)
# set up arrays for time t, velocity v, and position r
t = np.zeros(n)
v = np.zeros((n,2))
r = np.zeros((n,2))
# Initial conditions as compact 2-dimensional arrays
r0 = np.array([1.0,0.0])
v0 = np.array([0,8.9])
r[0] = r0
v[0] = v0
Fourpi2 = 4*pi*pi
# Start integrating using velocity verlet method
for i in range(n-1):
    # Set up the acceleration
    # Here you could have defined your own function for this
    rabs = sqrt(sum(r[i]*r[i]))
    a =  -Fourpi2*r[i]/(rabs**3)
    # update velocity, time and position using velocity-verlet method
    r[i+1] = r[i] + DeltaT*v[i] + a*DeltaT*DeltaT/2
    rabs = sqrt(sum(r[i+1]*r[i+1]))
    a_next = -Fourpi2*r[i+1]/(rabs**3)
    v[i+1] = v[i] + (DeltaT/2)*(a_next+a)
    
plt.title("velocity-verlet method")
plt.plot(r[:,0],r[:,1])
plt.xlabel("x position")
plt.ylabel("y position")
plt.show()
```

From my code it looks like the initial velocity needed to escape the sun is somewhere in the vicinity of 8.9 Au/year, as shown on the graph above.

To get an exact answer, we can choose the kinetic energy equal to the potential energy at 1Au from the sun, and solve for the velocity. Probably the easiest way is to use the well-known equation v = sqrt(2GM/r). Considering our units, this gives an escape velocity of sqrt(8*pi^2) = 8.886 Au/year, which agrees surprisingly well with what I predicted above!
