1a) In homework 5, I found the velocity verlet method to be more accurate than the euler-cromer method, so I'll use this. Explanation of some of my starting constants on handwritten paper, but essentially we can get G*M_sun = 4*pi^2 from assuming circular motion. We need a starting velocity of 2pi Au/year since we scaled our initial radius to be 1 Au (and clearly we need one revolution per year).

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
import statistics
%matplotlib inline

def velocityverlet(tfinal = 10, DeltaT = .001, beta=2):
    #set up arrays 
    n = ceil(tfinal/DeltaT)
    # set up arrays for time t, velocity v, and position r
    t = np.zeros(n)
    v = np.zeros((n,2))
    r = np.zeros((n,2))
    # Initial conditions as compact 2-dimensional arrays
    r0 = np.array([1.0,0.0])
    v0 = np.array([0,2*pi]) #We can alter this velocity a bit, but since we're starting at [1,0], we need it close to this to
                               #maintain a circular orbit
    r[0] = r0
    v[0] = v0
    Fourpi2 = 4*pi*pi
    # Start integrating using velocity verlet method
    for i in range(n-1):
        # Set up the acceleration
        # Here you could have defined your own function for this
        rabs = sqrt(sum(r[i]*r[i]))
        a =  -Fourpi2*r[i]/(rabs**(beta+1))
        # update velocity, time and position using velocity-verlet method
        r[i+1] = r[i] + DeltaT*v[i] + a*DeltaT*DeltaT/2
        rabs = sqrt(sum(r[i+1]*r[i+1]))
        a_next = -Fourpi2*r[i+1]/(rabs**(beta+1))
        v[i+1] = v[i] + (DeltaT/2)*(a_next+a)
        t[i+1] = t[i] + DeltaT
    return t,v,r
```

```python
t, v, r = velocityverlet()
plt.title("Plot for 1a")
plt.plot(r[:,0],r[:,1])
plt.xlabel("x position (Au)")
plt.ylabel("y position (Au)")
#plt.figure(figsize=(10,10))
#plt.set_aspect('equal')
plt.axis('equal')
plt.show()
```

By conservation of energy, this means kinetic energy must also be conserved. We could further justify conservation of kinetic energy by considering that you'd need a constant velocity to maintain a circular orbit, and a constant velocity (and constant mass) means constant kinetic energy.

1c) We stated in class that dA/dt = L/(2*mu) = 1/(2*mu)*magnitude(r x p) where p = mu*velocity. This means L/(2*mu) = magnitude(r x v)*1/2, or L = mu*magnitude(r x v) Note that r, p, and v are vectors and x represents the cross product. Because of this, we have that L is proportional to the magnitude of the cross product between the position r and the velocity v. We've numerically calculated r(t) and v(t), so if the cross product between those is constant it shows that L (the angular momentum) is constant, so angular momentum is conserved. To get an exact value, I'll multiply by the mass of the earth (mu).

```python
L = np.zeros(len(r))
for i in range(len(r)):
    L[i] = Mearth*np.cross(r[i]*Au,v[i]*Au/year)
plt.plot(t, L)
plt.title("Question 1c")
plt.xlabel("time (years)")
plt.ylabel("angular momentum (kg*m^2/s)")
plt.show()
print(statistics.mean(L))
```

We can see above that the angular momentum is constant, which makes sense. Since we're assuming a circular orbit, we always have constant magnitude of r and v, and they're always perpendicular to each other, so we'll always have the same value for our angular momentum. We can also see we get a value of around $2.66*10^{40} kg*m^2/s$, which agrees with the accepted value of ~$2.7*10^{40} kg*m^2/s$, as one would get using the mean radius from the sun and mean velocity of earth.

1e) In case you'd like to rerun all of these lines of code yourself, I don't want to mess with the initial condition from how I had them set in a-d, so I'll create a new function for this part. We know that total energy and angular momentum both need to be conserved for both beta=2 and beta=3 since we have no external forces/torques, but I think that the numerical analysis will be extremely sensitive to time step for beta=3 since we no longer expect an elliptical orbit (we have an unstable orbit). For this reason, I'll decrease the time step so that it's easier to tell that they're conserved (particularly angular momentum looked funny when I had too large of a time step).

```python
def velocityverlet2(tfinal = 10, DeltaT = .001, beta=2):
    #set up arrays 
    n = ceil(tfinal/DeltaT)
    # set up arrays for time t, velocity v, and position r
    t = np.zeros(n)
    v = np.zeros((n,2))
    r = np.zeros((n,2))
    # Initial conditions as compact 2-dimensional arrays
    r0 = np.array([1.0,0.0]) #Au
    v0 = np.array([0,5]) #Au/year
    r[0] = r0
    v[0] = v0
    Fourpi2 = 4*pi*pi
    # Start integrating using velocity verlet method
    for i in range(n-1):
        # Set up the acceleration
        # Here you could have defined your own function for this
        rabs = sqrt(sum(r[i]*r[i]))
        #rabs = np.linalg.norm(r[i])
        a =  -Fourpi2*r[i]/(rabs**(beta+1))
        # update velocity, time and position using velocity-verlet method
        r[i+1] = r[i] + DeltaT*v[i] + a*DeltaT*DeltaT/2
        rabs = sqrt(sum(r[i+1]*r[i+1]))
        a_next = -Fourpi2*r[i+1]/(rabs**(beta+1))
        v[i+1] = v[i] + (DeltaT/2)*(a_next+a)
        t[i+1] = t[i] + DeltaT
    return t,v,r
```

```python
#First, let's see what the orbit looks like
t, v, r = velocityverlet2()
plt.title("1e orbit beta=2")
plt.plot(r[:,0],r[:,1])
plt.xlabel("x position (Au)")
plt.ylabel("y position (Au)")
plt.axis('equal')
plt.show()
```

Sure enough, it's no longer circular. Since we decreased the velocity from 2*pi to 5, we have an ellipse along the x-axis. To show that total energy and angular momentum are conserved, I'll plot these and show that they're constant.

```python
Potential = -G*Msun*Mearth/(two_abs(r)*Au) #multiply r by Au to account for r measured in Au
Kinetic = .5*Mearth*(two_abs(v)*Au/year)**2
plt.title("1e energy beta=2")
plt.plot(t,Potential,label="potential energy")
plt.plot(t,Kinetic,label="kinetic energy")
plt.plot(t,Potential+Kinetic,label="total energy")
plt.xlabel("time (years)")
plt.ylabel("energy (J)")
plt.legend()
plt.show()
```

From the above plot, we can see that potential and kinetic energy both vary with time for an elliptical orbit, but the total energy remains constant.

```python
L = np.zeros(len(r))
for i in range(len(r)):
    L[i] = Mearth*np.cross(r[i]*Au,v[i]*Au/year)
plt.plot(t, L)
plt.title("1e angular momentum beta=2")
plt.xlabel("time (years)")
plt.ylabel("angular momentum (kg*m^2/s)")
plt.show()
```

The above plot shows that angular momentum is also conserved for the elliptical orbit with beta = 2.

```python
#I'm switching to 3D so  Ican use NASA data
def earthjupiter(tfinal = 10, DeltaT = .001):
    n = ceil(tfinal/DeltaT)
    t = np.zeros(n)
    
    earth_v = np.zeros((n,3))
    earth_r = np.zeros((n,3))
    jupiter_v = np.zeros((n,3))
    jupiter_r = np.zeros((n,3))
    
    # Initial conditions as compact 3-dimensional arrays
    earth_r0 = np.array([3.906262587512596E-01,8.996662220744706E-01,1.566155215319338E-04])
    earth_v0 = 365.25*np.array([-1.600480291707479E-02,6.881137189263926E-03,-6.493508848844465E-07])
    
    jupiter_r0 = np.array([3.661086547847372E+00,3.361387814148318E+00,-9.585416607358779E-02])
    jupiter_v0 = 365.25*np.array([-5.186614903007931E-03,5.915440748127593E-03,9.150523868195128E-05])
    
    earth_r[0] = earth_r0
    earth_v[0] = earth_v0
    jupiter_r[0] = jupiter_r0
    jupiter_v[0] = jupiter_v0
    
    Fourpi2 = 4*pi*pi
    # Start integrating using velocity verlet method
    for i in range(n-1):
        # Set up the acceleration
        # Here you could have defined your own function for this
        
        earth_a = -Fourpi2*earth_r[i]/(np.linalg.norm(earth_r[i])**3) - Fourpi2*(Mjupiter/Msun)*(earth_r[i]-jupiter_r[i])/np.linalg.norm(earth_r[i]-jupiter_r[i])**3
        jupiter_a = -Fourpi2*jupiter_r[i]/(np.linalg.norm(jupiter_r[i])**3) + Fourpi2*(Mjupiter/Msun)*(earth_r[i]-jupiter_r[i])/np.linalg.norm(earth_r[i]-jupiter_r[i])**3
        
        # update velocity, time and position using velocity-verlet method
        earth_r[i+1] = earth_r[i] + DeltaT*earth_v[i] + earth_a*DeltaT*DeltaT/2
        jupiter_r[i+1] = jupiter_r[i] + DeltaT*jupiter_v[i] + jupiter_a*DeltaT*DeltaT/2
        
        earth_a_next = -Fourpi2*earth_r[i+1]/(np.linalg.norm(earth_r[i+1])**3) - Fourpi2*(Mjupiter/Msun)*(earth_r[i+1]-jupiter_r[i+1])/np.linalg.norm(earth_r[i+1]-jupiter_r[i+1])**3
        jupiter_a_next = -Fourpi2*jupiter_r[i+1]/(np.linalg.norm(jupiter_r[i+1])**3) + Fourpi2*(Mjupiter/Msun)*(earth_r[i+1]-jupiter_r[i+1])/np.linalg.norm(earth_r[i+1]-jupiter_r[i+1])**3
        
        earth_v[i+1] = earth_v[i] + (DeltaT/2)*(earth_a_next+earth_a)
        jupiter_v[i+1] = jupiter_v[i] + (DeltaT/2)*(jupiter_a_next+jupiter_a)
        
        t[i+1] = t[i] + DeltaT
    return t,earth_v,earth_r,jupiter_v,jupiter_r
```

```python
#2c) moving on to the full solar system
def solarsystem(tfinal = 50, DeltaT = .001):
    n = ceil(tfinal/DeltaT)
    t = np.zeros(n)
    
    earth_v = np.zeros((n,3))
    earth_r = np.zeros((n,3))
    jupiter_v = np.zeros((n,3))
    jupiter_r = np.zeros((n,3))
    mars_v = np.zeros((n,3))
    mars_r = np.zeros((n,3))
    venus_v = np.zeros((n,3))
    venus_r = np.zeros((n,3))
    saturn_v = np.zeros((n,3))
    saturn_r = np.zeros((n,3))
    mercury_v = np.zeros((n,3))
    mercury_r = np.zeros((n,3))
    uranus_v = np.zeros((n,3))
    uranus_r = np.zeros((n,3))
    neptune_v = np.zeros((n,3))
    neptune_r = np.zeros((n,3))
    
    # Initial conditions as compact 3-dimensional arrays
    earth_r0 = np.array([3.906262587512596E-01,8.996662220744706E-01,1.566155215319338E-04])
    earth_v0 = 365.25*np.array([-1.600480291707479E-02,6.881137189263926E-03,-6.493508848844465E-07])  
    jupiter_r0 = np.array([3.661086547847372E+00,3.361387814148318E+00,-9.585416607358779E-02])
    jupiter_v0 = 365.25*np.array([-5.186614903007931E-03,5.915440748127593E-03,9.150523868195128E-05])
    mars_r0 = np.array([-7.500598055702374E-01,-1.333271651172154E+00,-9.474935513331478E-03])
    mars_v0 = 365.25*np.array([1.275316544951573E-02,-5.623747515463687E-03,-4.304561886499535E-04])
    venus_r0 = np.array([-4.933483385766508E-01,5.259935650564580E-01,3.546985873534401E-02])
    venus_v0 = 365.25*np.array([-1.496792821945952E-02,-1.379214325696699E-02,6.745757820577678E-04])
    saturn_r0 = np.array([8.926427060617058E+00,-3.888158283640868E+00,-2.877982347424594E-01])
    saturn_v0 = 365.25*np.array([1.916444749051573E-03,5.103160074489122E-03,-1.647209373518789E-04])
    mercury_r0 = np.array([3.208344702418472E-01,-2.278159559102075E-01,-4.836105008944021E-02])
    mercury_v0 = 365.25*np.array([1.035996473151648E-02,2.450729008662945E-02,1.053485554398226E-03])
    uranus_r0 = np.array([1.236498380010862E+01,1.522192708157192E+01,-1.036559263081453E-01])
    uranus_v0 = 365.25*np.array([-3.081601578143489E-03,2.296558441414854E-03,4.845888162893780E-05])
    neptune_r0 = np.array([2.982930011724746E+01,-1.897820838202275E+00,-6.483648286736278E-01])
    neptune_v0 = 365.25*np.array([1.787103984625802E-04,3.151712844925773E-03,-6.886487442265150E-05])
    
    earth_r[0] = earth_r0
    earth_v[0] = earth_v0
    jupiter_r[0] = jupiter_r0
    jupiter_v[0] = jupiter_v0
    mars_r[0] = mars_r0
    mars_v[0] = mars_v0
    venus_r[0] = venus_r0
    venus_v[0] = venus_v0
    saturn_r[0] = saturn_r0
    saturn_v[0] = saturn_v0
    mercury_r[0] = mercury_r0
    mercury_v[0] = mercury_v0
    uranus_r[0] = uranus_r0
    uranus_v[0] = uranus_v0
    neptune_r[0] = neptune_r0
    neptune_v[0] = neptune_v0
    
    Fourpi2 = 4*pi*pi
    # Start integrating using velocity verlet method
    for i in range(n-1):
        #all of our forces (each gets used twice, once for each planet involved)
        earth_jupiter = Fourpi2*(Mjupiter/Msun)*(earth_r[i]-jupiter_r[i])/np.linalg.norm(earth_r[i]-jupiter_r[i])**3
        earth_mars = Fourpi2*(Mmars/Msun)*(earth_r[i]-mars_r[i])/np.linalg.norm(earth_r[i]-mars_r[i])**3
        jupiter_mars = Fourpi2*(Mmars/Msun)*(jupiter_r[i]-mars_r[i])/np.linalg.norm(jupiter_r[i]-mars_r[i])**3
        earth_venus = Fourpi2*(Mvenus/Msun)*(earth_r[i]-venus_r[i])/np.linalg.norm(earth_r[i]-venus_r[i])**3
        jupiter_venus = Fourpi2*(Mvenus/Msun)*(jupiter_r[i]-venus_r[i])/np.linalg.norm(jupiter_r[i]-venus_r[i])**3
        mars_venus = Fourpi2*(Mvenus/Msun)*(mars_r[i]-venus_r[i])/np.linalg.norm(mars_r[i]-venus_r[i])**3
        earth_saturn = Fourpi2*(Msaturn/Msun)*(earth_r[i]-saturn_r[i])/np.linalg.norm(earth_r[i]-saturn_r[i])**3
        jupiter_saturn = Fourpi2*(Msaturn/Msun)*(jupiter_r[i]-saturn_r[i])/np.linalg.norm(jupiter_r[i]-saturn_r[i])**3
        mars_saturn = Fourpi2*(Msaturn/Msun)*(mars_r[i]-saturn_r[i])/np.linalg.norm(mars_r[i]-saturn_r[i])**3
        venus_saturn = Fourpi2*(Msaturn/Msun)*(venus_r[i]-saturn_r[i])/np.linalg.norm(venus_r[i]-saturn_r[i])**3
        earth_mercury = Fourpi2*(Mmercury/Msun)*(earth_r[i]-mercury_r[i])/np.linalg.norm(earth_r[i]-mercury_r[i])**3
        jupiter_mercury = Fourpi2*(Mmercury/Msun)*(jupiter_r[i]-mercury_r[i])/np.linalg.norm(jupiter_r[i]-mercury_r[i])**3
        mars_mercury = Fourpi2*(Mmercury/Msun)*(mars_r[i]-mercury_r[i])/np.linalg.norm(mars_r[i]-mercury_r[i])**3
        venus_mercury = Fourpi2*(Mmercury/Msun)*(venus_r[i]-mercury_r[i])/np.linalg.norm(venus_r[i]-mercury_r[i])**3
        saturn_mercury = Fourpi2*(Mmercury/Msun)*(saturn_r[i]-mercury_r[i])/np.linalg.norm(saturn_r[i]-mercury_r[i])**3
        earth_uranus = Fourpi2*(Muranus/Msun)*(earth_r[i]-uranus_r[i])/np.linalg.norm(earth_r[i]-uranus_r[i])**3
        jupiter_uranus = Fourpi2*(Muranus/Msun)*(jupiter_r[i]-uranus_r[i])/np.linalg.norm(jupiter_r[i]-uranus_r[i])**3
        mars_uranus = Fourpi2*(Muranus/Msun)*(mars_r[i]-uranus_r[i])/np.linalg.norm(mars_r[i]-uranus_r[i])**3
        venus_uranus = Fourpi2*(Muranus/Msun)*(venus_r[i]-uranus_r[i])/np.linalg.norm(venus_r[i]-uranus_r[i])**3
        saturn_uranus = Fourpi2*(Muranus/Msun)*(saturn_r[i]-uranus_r[i])/np.linalg.norm(saturn_r[i]-uranus_r[i])**3
        mercury_uranus = Fourpi2*(Muranus/Msun)*(mercury_r[i]-uranus_r[i])/np.linalg.norm(mercury_r[i]-uranus_r[i])**3
        earth_neptune = Fourpi2*(Mneptune/Msun)*(earth_r[i]-neptune_r[i])/np.linalg.norm(earth_r[i]-neptune_r[i])**3
        jupiter_neptune = Fourpi2*(Mneptune/Msun)*(jupiter_r[i]-neptune_r[i])/np.linalg.norm(jupiter_r[i]-neptune_r[i])**3
        mars_neptune = Fourpi2*(Mneptune/Msun)*(mars_r[i]-neptune_r[i])/np.linalg.norm(mars_r[i]-neptune_r[i])**3
        venus_neptune = Fourpi2*(Mneptune/Msun)*(venus_r[i]-neptune_r[i])/np.linalg.norm(venus_r[i]-neptune_r[i])**3
        saturn_neptune = Fourpi2*(Mneptune/Msun)*(saturn_r[i]-neptune_r[i])/np.linalg.norm(saturn_r[i]-neptune_r[i])**3
        mercury_neptune = Fourpi2*(Mneptune/Msun)*(mercury_r[i]-neptune_r[i])/np.linalg.norm(mercury_r[i]-neptune_r[i])**3
        uranus_neptune = Fourpi2*(Mneptune/Msun)*(uranus_r[i]-neptune_r[i])/np.linalg.norm(uranus_r[i]-neptune_r[i])**3
        
        #planet accelerations at i
        earth_a = -Fourpi2*earth_r[i]/(np.linalg.norm(earth_r[i])**3)-earth_jupiter-earth_mars-earth_venus-earth_saturn-earth_mercury-earth_uranus-earth_neptune
        jupiter_a = -Fourpi2*jupiter_r[i]/(np.linalg.norm(jupiter_r[i])**3)+earth_jupiter-jupiter_mars-jupiter_venus-jupiter_saturn-jupiter_mercury-jupiter_uranus-jupiter_neptune
        mars_a = -Fourpi2*mars_r[i]/(np.linalg.norm(mars_r[i])**3)+earth_mars+jupiter_mars-mars_venus-mars_saturn-mars_mercury-mars_uranus-mars_neptune
        venus_a = -Fourpi2*venus_r[i]/(np.linalg.norm(venus_r[i])**3)+earth_venus+jupiter_venus+mars_venus-venus_saturn-venus_mercury-venus_uranus-venus_neptune
        saturn_a = -Fourpi2*saturn_r[i]/(np.linalg.norm(saturn_r[i])**3)+earth_saturn+jupiter_saturn+mars_saturn+venus_saturn-saturn_mercury-saturn_uranus-saturn_neptune
        mercury_a = -Fourpi2*mercury_r[i]/(np.linalg.norm(mercury_r[i])**3)+earth_mercury+jupiter_mercury+mars_mercury+venus_mercury+saturn_mercury-mercury_uranus-mercury_neptune
        uranus_a = -Fourpi2*uranus_r[i]/(np.linalg.norm(uranus_r[i])**3)+earth_uranus+jupiter_uranus+mars_uranus+venus_uranus+saturn_uranus+mercury_uranus-uranus_neptune
        neptune_a = -Fourpi2*neptune_r[i]/(np.linalg.norm(neptune_r[i])**3)+earth_neptune+jupiter_neptune+mars_neptune+venus_neptune+saturn_neptune+mercury_neptune+uranus_neptune
        
        # update using velocity-verlet method
        #updating positions
        earth_r[i+1] = earth_r[i] + DeltaT*earth_v[i] + earth_a*DeltaT*DeltaT/2
        jupiter_r[i+1] = jupiter_r[i] + DeltaT*jupiter_v[i] + jupiter_a*DeltaT*DeltaT/2
        mars_r[i+1] = mars_r[i] + DeltaT*mars_v[i] + mars_a*DeltaT*DeltaT/2
        venus_r[i+1] = venus_r[i] + DeltaT*venus_v[i] + venus_a*DeltaT*DeltaT/2
        saturn_r[i+1] = saturn_r[i] + DeltaT*saturn_v[i] + saturn_a*DeltaT*DeltaT/2
        mercury_r[i+1] = mercury_r[i] + DeltaT*mercury_v[i] + mercury_a*DeltaT*DeltaT/2
        uranus_r[i+1] = uranus_r[i] + DeltaT*uranus_v[i] + uranus_a*DeltaT*DeltaT/2
        neptune_r[i+1] = neptune_r[i] + DeltaT*neptune_v[i] + neptune_a*DeltaT*DeltaT/2
        
        #redefine these forces for i+1
        earth_jupiter = Fourpi2*(Mjupiter/Msun)*(earth_r[i+1]-jupiter_r[i+1])/np.linalg.norm(earth_r[i+1]-jupiter_r[i+1])**3
        earth_mars = Fourpi2*(Mmars/Msun)*(earth_r[i+1]-mars_r[i+1])/np.linalg.norm(earth_r[i+1]-mars_r[i+1])**3
        jupiter_mars = Fourpi2*(Mmars/Msun)*(jupiter_r[i+1]-mars_r[i+1])/np.linalg.norm(jupiter_r[i+1]-mars_r[i+1])**3
        earth_venus = Fourpi2*(Mvenus/Msun)*(earth_r[i+1]-venus_r[i+1])/np.linalg.norm(earth_r[i+1]-venus_r[i+1])**3
        jupiter_venus = Fourpi2*(Mvenus/Msun)*(jupiter_r[i+1]-venus_r[i+1])/np.linalg.norm(jupiter_r[i+1]-venus_r[i+1])**3
        mars_venus = Fourpi2*(Mvenus/Msun)*(mars_r[i+1]-venus_r[i+1])/np.linalg.norm(mars_r[i+1]-venus_r[i+1])**3
        earth_saturn = Fourpi2*(Msaturn/Msun)*(earth_r[i+1]-saturn_r[i+1])/np.linalg.norm(earth_r[i+1]-saturn_r[i+1])**3
        jupiter_saturn = Fourpi2*(Msaturn/Msun)*(jupiter_r[i+1]-saturn_r[i+1])/np.linalg.norm(jupiter_r[i+1]-saturn_r[i+1])**3
        mars_saturn = Fourpi2*(Msaturn/Msun)*(mars_r[i+1]-saturn_r[i+1])/np.linalg.norm(mars_r[i+1]-saturn_r[i+1])**3
        venus_saturn = Fourpi2*(Msaturn/Msun)*(venus_r[i+1]-saturn_r[i+1])/np.linalg.norm(venus_r[i+1]-saturn_r[i+1])**3
        earth_mercury = Fourpi2*(Mmercury/Msun)*(earth_r[i+1]-mercury_r[i+1])/np.linalg.norm(earth_r[i+1]-mercury_r[i+1])**3
        jupiter_mercury = Fourpi2*(Mmercury/Msun)*(jupiter_r[i+1]-mercury_r[i+1])/np.linalg.norm(jupiter_r[i+1]-mercury_r[i+1])**3
        mars_mercury = Fourpi2*(Mmercury/Msun)*(mars_r[i+1]-mercury_r[i+1])/np.linalg.norm(mars_r[i+1]-mercury_r[i+1])**3
        venus_mercury = Fourpi2*(Mmercury/Msun)*(venus_r[i+1]-mercury_r[i+1])/np.linalg.norm(venus_r[i+1]-mercury_r[i+1])**3
        saturn_mercury = Fourpi2*(Mmercury/Msun)*(saturn_r[i+1]-mercury_r[i+1])/np.linalg.norm(saturn_r[i+1]-mercury_r[i+1])**3
        earth_uranus = Fourpi2*(Muranus/Msun)*(earth_r[i+1]-uranus_r[i+1])/np.linalg.norm(earth_r[i+1]-uranus_r[i+1])**3
        jupiter_uranus = Fourpi2*(Muranus/Msun)*(jupiter_r[i+1]-uranus_r[i+1])/np.linalg.norm(jupiter_r[i+1]-uranus_r[i+1])**3
        mars_uranus = Fourpi2*(Muranus/Msun)*(mars_r[i+1]-uranus_r[i+1])/np.linalg.norm(mars_r[i+1]-uranus_r[i+1])**3
        venus_uranus = Fourpi2*(Muranus/Msun)*(venus_r[i+1]-uranus_r[i+1])/np.linalg.norm(venus_r[i+1]-uranus_r[i+1])**3
        saturn_uranus = Fourpi2*(Muranus/Msun)*(saturn_r[i+1]-uranus_r[i+1])/np.linalg.norm(saturn_r[i+1]-uranus_r[i+1])**3
        mercury_uranus = Fourpi2*(Muranus/Msun)*(mercury_r[i+1]-uranus_r[i+1])/np.linalg.norm(mercury_r[i+1]-uranus_r[i+1])**3
        earth_neptune = Fourpi2*(Mneptune/Msun)*(earth_r[i+1]-neptune_r[i+1])/np.linalg.norm(earth_r[i+1]-neptune_r[i+1])**3
        jupiter_neptune = Fourpi2*(Mneptune/Msun)*(jupiter_r[i+1]-neptune_r[i+1])/np.linalg.norm(jupiter_r[i+1]-neptune_r[i+1])**3
        mars_neptune = Fourpi2*(Mneptune/Msun)*(mars_r[i+1]-neptune_r[i+1])/np.linalg.norm(mars_r[i+1]-neptune_r[i+1])**3
        venus_neptune = Fourpi2*(Mneptune/Msun)*(venus_r[i+1]-neptune_r[i+1])/np.linalg.norm(venus_r[i+1]-neptune_r[i+1])**3
        saturn_neptune = Fourpi2*(Mneptune/Msun)*(saturn_r[i+1]-neptune_r[i+1])/np.linalg.norm(saturn_r[i+1]-neptune_r[i+1])**3
        mercury_neptune = Fourpi2*(Mneptune/Msun)*(mercury_r[i+1]-neptune_r[i+1])/np.linalg.norm(mercury_r[i+1]-neptune_r[i+1])**3
        uranus_neptune = Fourpi2*(Mneptune/Msun)*(uranus_r[i+1]-neptune_r[i+1])/np.linalg.norm(uranus_r[i+1]-neptune_r[i+1])**3
        
        #accelerations for i+1
        earth_a_next = -Fourpi2*earth_r[i+1]/(np.linalg.norm(earth_r[i+1])**3)-earth_jupiter-earth_mars-earth_venus-earth_saturn-earth_mercury-earth_uranus-earth_neptune
        jupiter_a_next = -Fourpi2*jupiter_r[i+1]/(np.linalg.norm(jupiter_r[i+1])**3)+earth_jupiter-jupiter_mars-jupiter_venus-jupiter_saturn-jupiter_mercury-jupiter_uranus-jupiter_neptune
        mars_a_next = -Fourpi2*mars_r[i+1]/(np.linalg.norm(mars_r[i+1])**3)+earth_mars+jupiter_mars-mars_venus-mars_saturn-mars_mercury-mars_uranus-mars_neptune
        venus_a_next = -Fourpi2*venus_r[i+1]/(np.linalg.norm(venus_r[i+1])**3)+earth_venus+jupiter_venus+mars_venus-venus_saturn-venus_mercury-venus_uranus-venus_neptune
        saturn_a_next = -Fourpi2*saturn_r[i+1]/(np.linalg.norm(saturn_r[i+1])**3)+earth_saturn+jupiter_saturn+mars_saturn+venus_saturn-saturn_mercury-saturn_uranus-saturn_neptune
        mercury_a_next = -Fourpi2*mercury_r[i+1]/(np.linalg.norm(mercury_r[i+1])**3)+earth_mercury+jupiter_mercury+mars_mercury+venus_mercury+saturn_mercury-mercury_uranus-mercury_neptune
        uranus_a_next = -Fourpi2*uranus_r[i+1]/(np.linalg.norm(uranus_r[i+1])**3)+earth_uranus+jupiter_uranus+mars_uranus+venus_uranus+saturn_uranus+mercury_uranus-uranus_neptune
        neptune_a_next = -Fourpi2*neptune_r[i+1]/(np.linalg.norm(neptune_r[i+1])**3)+earth_neptune+jupiter_neptune+mars_neptune+venus_neptune+saturn_neptune+mercury_neptune+uranus_neptune
        
        #updating velocities
        earth_v[i+1] = earth_v[i] + (DeltaT/2)*(earth_a_next+earth_a)
        jupiter_v[i+1] = jupiter_v[i] + (DeltaT/2)*(jupiter_a_next+jupiter_a)
        mars_v[i+1] = mars_v[i] + (DeltaT/2)*(mars_a_next+mars_a)
        venus_v[i+1] = venus_v[i] + (DeltaT/2)*(venus_a_next+venus_a)
        saturn_v[i+1] = saturn_v[i] + (DeltaT/2)*(saturn_a_next+saturn_a)
        mercury_v[i+1] = mercury_v[i] + (DeltaT/2)*(mercury_a_next+mercury_a)
        uranus_v[i+1] = uranus_v[i] + (DeltaT/2)*(uranus_a_next+uranus_a)
        neptune_v[i+1] = neptune_v[i] + (DeltaT/2)*(neptune_a_next+neptune_a)
        
        t[i+1] = t[i] + DeltaT
    return t,earth_v,earth_r,jupiter_v,jupiter_r,mars_v,mars_r,venus_v,venus_r,saturn_v,saturn_r,mercury_v,mercury_r,uranus_v,uranus_r,neptune_v,neptune_r
```

I'm actually going to change so that I haven't scaled my equations. With how I had things set up before, I ran into an issue with Bertrand's Theorem. Esssentially the problem was that I had set things up to guarantee that the centripetal force is equal to the central force. By Bertrand's Theorem, this condition alone guarantees a circular orbit for all central forces, so I was just getting circular orbits. I'm assuming this isn't what was wanted, so I'm going to change my conditions slightly.

```python
def velocityverlet_unscaled(DeltaT = .001, beta=2):
    #set up arrays 
    tfinal = 10 # in years
    n = ceil(tfinal/DeltaT)
    # set up arrays for time t, velocity v, and position r
    t = np.zeros(n)
    v = np.zeros((n,2))
    r = np.zeros((n,2))
    # Initial conditions as compact 2-dimensional arrays
    r0 = np.array([Au,0.0])
    v0 = np.array([0,2*Au*pi]) #We can alter this velocity a bit, but since we're starting at [1,0], we need it close to this to
                               #maintain a circular orbit
    r[0] = r0
    v[0] = v0
    #Fourpi2 = 4*pi*pi
    # Start integrating using velocity verlet method
    for i in range(n-1):
        rabs = sqrt(sum(r[i]*r[i]))
        a =  -G*Msun*r[i]/(rabs**(beta+1))
        # update velocity, time and position using velocity-verlet method
        r[i+1] = r[i] + DeltaT*v[i] + a*DeltaT*DeltaT/2
        rabs = sqrt(sum(r[i+1]*r[i+1]))
        a_next = -G*Msun*r[i+1]/(rabs**(beta+1))
        v[i+1] = v[i] + (DeltaT/2)*(a_next+a)
        t[i+1] = t[i] + DeltaT
    return t,v,r
```
