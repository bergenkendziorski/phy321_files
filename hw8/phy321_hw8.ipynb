### Exercise 1 (10pt), Equations of motion and more

The equations of motion in the center-of-mass frame in two dimensions with $x=r\cos{(\phi)}$ and $y=r\sin{(\phi)}$ and
$r\in [0,\infty)$, $\phi\in [0,2\pi]$ and $r=\sqrt{x^2+y^2}$ are given by

$$
\mu \ddot{r}=-\frac{dV(r)}{dr}+\mu r\dot{\phi}^2,
$$

and

$$
\dot{\phi}=\frac{L}{\mu r^2}.
$$

Here $V(r)$ is any central force which depends only on the relative coordinate.
* 1a (5pt) Show that you can rewrite the radial equation in terms of an effective potential $V_{\mathrm{eff}}(r)=V(r)+L^2/(2\mu r^2)$. 

* 1b (5pt) Write out the final differential equation for the radial degrees of freedom when we specify that $V(r)=-\alpha/r$.  Plot the effective potential. You can choose values for $\alpha$ and $L$ and discuss (see Taylor section 8.4 and example 8.2) the physics of the system for two energies, one larger than zero and one smaller than zero. This is similar to what you did in the first midterm, except that the potential is different.

```python
#1b plot)
#differential equation shown on written paper
import matplotlib.pyplot as plt
import numpy as np

r = np.arange(.35,10,.001)
alpha = 1
L = 1
mu = 1
Veff = -alpha/r + L**2/(2*mu*r**2) #I'm just choosing to set alpha = L = mu = 1 for simplicity
plt.title("question 1b plot")
plt.xlabel("r")
plt.ylabel("Veff")
plt.plot(r,Veff)
```

With energy less than zero, we can see that an object would stay stuck in the potential well. If it had energy exactly the minimum of the above function, then it would have a circular orbit. If it had energy greater than this but less than 0 it would have elliptical motion, as shown by going back and forth between the two different r values that have that Energy on the above graph.
With energy greater than 0, the particle would have enough energy to escape the potential well, and would therefore have parabolic/hyperparabolic motion (essentially escape the orbit, and have an unbounded orbit).

### Exercise 3 (40pt), Numerical Solution of the Harmonic Oscillator

Using the code we developed in homework 5 or the first midterm (part b), we can solve the above harmonic oscillator problem in two dimensions using our code from this homework. We need however to change the acceleration from the gravitational force to the one given by the harmonic oscillator potential.

The code is given here for the Velocity-Verlet algorithm, with obvious elements to fill in.

* 3a (20pt) Use for example the above code to set up the acceleration and use the initial conditions fixed by for example $r_{\rm min}$ from exercise 2. Which value should the initial velocity take if you place yourself at $r_{\rm min}$ and you require a circular motion? Hint: see the first midterm, part 2. There you used the centripetal acceleration.  

Instead of solving the equations in the cartesian frame we will now rewrite the above code in terms of the radial degrees of freedom only. Our differential equation is now

$$
\mu \ddot{r}=-\frac{dV(r)}{dr}+\mu r\dot{\phi}^2,
$$

and

$$
\dot{\phi}=\frac{L}{\mu r^2}.
$$

3a) We found rmin in part 2a. So we can set x initial to rmin and y initial to 0. To get circular motion, we need initial velocity in the y-direction with magnitude so that centripetal force is equal to the force from the potential. Doing this gives (analysis on magnitude) mv^2/r = kr, so mv^2 = kr^2, and v = sqrt(k/m)*r. Since in 2c we showed omega0 = sqrt(k/m), we can say the magnitude of our initial velocity must be equal to omega0 *r.

```python
#3a code
%matplotlib inline

# Common imports
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt

DeltaT = 0.001
#set up arrays 
tfinal = 10.0
n = ceil(tfinal/DeltaT)
# set up arrays for t, v, and r
t = np.zeros(n)
v = np.zeros((n,2))
r = np.zeros((n,2))

#constants. I'm just setting things to one for ease
L = 1.0
k = 1.0   # spring constant
m = 1.0   # mass
omega02 = k/m
r_min = (L*L/(k*m))**(1/4) #should be just 1.0

# Initial conditions as compact 2-dimensional arrays
r0 = np.array([r_min,0])
v0 = np.array([0.0,np.sqrt(omega02)*r_min])
r[0] = r0
v[0] = v0

# Start integrating using the Velocity-Verlet  method
for i in range(n-1):
    a =  -r[i]*omega02 #this acceleration is fine for now
    # update velocity, time and position using the Velocity-Verlet method
    r[i+1] = r[i] + DeltaT*v[i]+0.5*(DeltaT**2)*a
    # new accelerationfor the Verlet method
    anew = -r[i+1]*omega02  
    v[i+1] = v[i] + 0.5*DeltaT*(a+anew)
    t[i+1] = t[i] + DeltaT
# Plot position as function of time    
fig, ax = plt.subplots()
ax.set_xlabel('t[s]')
ax.set_ylabel('position (m)')
ax.plot(t,r[:,0],label = "x position")
ax.plot(t,r[:,1],label = "y position")
ax.plot(t,np.sqrt(r[:,0]**2 + r[:,1]**2),label = "radial distance")
fig.tight_layout()
plt.legend()
plt.title("Question 3a")
plt.show()

fig, ax = plt.subplots()
ax.set_xlabel('t[s]')
ax.set_ylabel('energy (J)')
ax.plot(t,.5*m*(v[:,0]**2 + v[:,1]**2),label = "kinetic energy") #KE = .5mv^2 = .5*m*(v_x^2 + v_y^2)
ax.plot(t,.5*k*(r[:,0]**2 + r[:,1]**2) + .5*m*(v[:,0]**2 + v[:,1]**2),label = "total energy")
fig.tight_layout()
plt.legend()
plt.title("Question 3a")
plt.show()
```

From the above we can see that things are working properly. We have out of phase oscillatory motion in x and y positions, radial distance stays at the constant 1 (which is what we set our initial x-position to), and our kinetic energy and total energy stay constant.

3b) I believe this is just asking for me to do the same thing in polar coordinates. The general process stays the same, but now we need to add in the additional effective force term/centrifugal barrier (think added effective potential term in the total potential).

Our potential has an added L^2/(2*m*r^2). For the acceleration, we have an additional (from negative derivative with respect to r of this added potential) F=m*a=L^2/(m*r^3), or a = L^2/(m^2*r^3), so this needs to be added to our acceleration. Also add in the additional potential L*L/(2*m*r*r) when calculating potential energy

```python
#3b code
%matplotlib inline

# Common imports
import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt

DeltaT = 0.001
#set up arrays 
tfinal = 10.0
n = ceil(tfinal/DeltaT)
# set up arrays for t, v, and r
t = np.zeros(n)
v = np.zeros(n) #now one dimensional
r = np.zeros(n) #now one dimensional

#constants. I'm just setting things to one for ease
L = 1.0
k = 1.0   # spring constant
m = 1.0   # mass
omega02 = k/m
r_min = (L*L/(k*m))**(1/4) #should be just 1.0

# Initial conditions
r0 = r_min
v0 = 0
r[0] = r0
v[0] = v0

# Start integrating using the Velocity-Verlet  method
for i in range(n-1):
    a =  -r[i]*omega02 + (L*L)/(m*m*(r[i]**3))
    # update velocity, time and position using the Velocity-Verlet method
    r[i+1] = r[i] + DeltaT*v[i]+0.5*(DeltaT**2)*a
    # new accelerationfor the Verlet method
    anew = -r[i+1]*omega02 + (L*L)/(m*m*(r[i+1]**3))
    v[i+1] = v[i] + 0.5*DeltaT*(a+anew)
    t[i+1] = t[i] + DeltaT
# Plot position as function of time 
KineticEn = .5*m*v**2
PotentialEn = .5*k*r**2 + L*L/(2*m*r*r)

plt.subplot(2,1,1)
plt.title("Question 3b")
plt.xlabel('t[s]')
plt.ylabel('radial distance (m)')
plt.plot(t,r)

plt.subplot(2,1,2)
plt.plot(t,KineticEn+PotentialEn)
plt.xlabel('t[s]')
plt.ylabel('total energy (J)')
#fig.tight_layout()
plt.show()
```

From this, we can see that it does agree perfectly with what we had in 3a. We still have a constant radius of 1m and a total energy of 1 Joule.

### Bonus exercise, from [Julie Butler's notes](https://github.com/mhjensen/Physics321/blob/master/doc/Codes/ClassicalMechanicsSolvers/WritingReusableCode.ipynb)

This bonus exercise gives you an additional score of **20pt**.

Make a class called Solvers which contains the following:
1. An initializer

2. A method that implements Euler's method in 2D

3. A method that implements Euler-Cromer's method in 2D

4. A method that implements Velocity-Verlet in 2D

5. A method that plots position versus time results from the solver

oAt least three class level variables

The class may also contain any other methods or class level variables that you think are needed.

Make then a function (outside of the class Solver) that solves the Earth-Sun problem using the Solvers class (Exercise 6 from Homework 5).

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
%matplotlib inline

class Solvers:
    Fourpi2 = 4*pi*pi
    tfinal = 10
    DeltaT = .001
    
    def __init__(self):
        print("created")

    def setup(self):
        self.n = ceil(self.tfinal/self.DeltaT)
        self.t = np.zeros(self.n)
        self.v = np.zeros((self.n,2))
        self.r = np.zeros((self.n,2))
        # Initial conditions as compact 2-dimensional arrays
        r0 = np.array([1.0,0.0])
        v0 = np.array([0,2*pi]) #We can alter this velocity a bit, but since we're starting at [1,0], we need it close to this to
                                   #maintain a circular orbit
        self.r[0] = r0
        self.v[0] = v0
        
    def euler(self):
        for i in range(self.n-1):
            rabs = sqrt(sum(self.r[i]*self.r[i]))
            a =  -1*self.Fourpi2*self.r[i]/(rabs**3)
            # update velocity, time and position using Euler's forward method
            self.v[i+1] = self.v[i] + self.DeltaT*a
            self.r[i+1] = self.r[i] + self.DeltaT*self.v[i]
            self.t[i+1] = self.t[i] + self.DeltaT
            
    def eulercromer(self):
        # Start integrating using Euler-Cromer's method
        for i in range(self.n-1):
            # Set up the acceleration
            # Here you could have defined your own function for this
            rabs = sqrt(sum(self.r[i]*self.r[i]))
            a =  -1*self.Fourpi2*self.r[i]/(rabs**3)
            # update velocity, time and position using Euler's forward method
            self.v[i+1] = self.v[i] + self.DeltaT*a
            self.r[i+1] = self.r[i] + self.DeltaT*self.v[i+1]
            self.t[i+1] = self.t[i] + self.DeltaT
            
    def velocityverlet(self):
        # Start integrating using velocity verlet method
        for i in range(self.n-1):
            # Set up the acceleration
            # Here you could have defined your own function for this
            rabs = sqrt(sum(self.r[i]*self.r[i]))
            a =  -1*self.Fourpi2*self.r[i]/(rabs**3)
            # update velocity, time and position using velocity-verlet method
            self.r[i+1] = self.r[i] + self.DeltaT*self.v[i] + a*self.DeltaT*self.DeltaT/2
            rabs = sqrt(sum(self.r[i+1]*self.r[i+1]))
            a_next = -1*self.Fourpi2*self.r[i+1]/(rabs**3)
            self.v[i+1] = self.v[i] + (self.DeltaT/2)*(a_next+a)
            self.t[i+1] = self.t[i] + self.DeltaT
        
    def plotting(self):
        plt.subplot(2,1,1)
        plt.plot(self.r[:,0],self.r[:,1])
        plt.xlabel("x position")
        plt.ylabel("y position")
        
        plt.subplot(2,1,2)
        plt.plot(self.t,self.r[:,0],label="x position")
        plt.plot(self.t,self.r[:,1],label="y position")
        plt.xlabel("time")
        plt.ylabel("position")
        plt.legend()
        plt.tight_layout()
        plt.show()
        
```

```python
#Now I'll just make a function to run it
def solve_earth_sun(method = "euler"):
    if(method == "euler"):
        print("solving with euler's method")
        euler_tester = Solvers()
        euler_tester.setup()
        euler_tester.euler()
        euler_tester.plotting()
    elif(method == "eulercromer"):
        print("solving using euler-cromer")
        eulercromer_tester = Solvers()
        eulercromer_tester.setup()
        eulercromer_tester.eulercromer()
        eulercromer_tester.plotting()
    elif(method == "velocityverlet"):
        print("solving using velocity-verlet")
        velocityverlet_tester = Solvers()
        velocityverlet_tester.setup()
        velocityverlet_tester.velocityverlet()
        velocityverlet_tester.plotting()
    else:
        print("Method type invalid")
```

```python
solve_earth_sun("euler")
```
